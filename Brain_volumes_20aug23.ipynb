```python
regions = list(cm_vols_ttest.columns[1:15])
```

```python
case_vols = cm_vols_ttest.loc[cm_vols_ttest['Case/control']=='Case']
control_vols = cm_vols_ttest.loc[cm_vols_ttest['Case/control']=='Control']
```

```python
shapiroes = []
levenes = []
for region in regions:
    diff = case_vols[region].values - control_vols[region].values
    shapiro = [stats.shapiro(diff)[1]]
    levene = [stats.levene(case_vols[region].values, control_vols[region].values, center='mean')[1]]
    shapiro = pd.DataFrame(shapiro)
    shapiro['region'] = region
    levene = pd.DataFrame(levene)
    levene['region'] = region
    shapiroes.append(shapiro)
    levenes.append(levene)
```

```python
shapiroes = pd.concat(shapiroes)
```

```python
levenes = pd.concat(levenes)
```

```python
### t-test
```

```python
t_s = []
p_s = []
for region in regions:
    case_array = np.array(case_vols[region])
    control_array = np.array(control_vols[region])
    t = [stats.ttest_ind(case_array, control_array)[0]]
    t = pd.DataFrame(t)
    t['region']=region
    p = [stats.ttest_ind(case_array, control_array)[1]]
    p = pd.DataFrame(p)
    p['region']=region
    t_s.append(t)
    p_s.append(p)
```

```python
t_s = pd.concat(t_s)
p_s = pd.concat(p_s)
```

```python
p_s_array = np.array(p_s[0])
```

```python
bonferroni_ps = multipletests(p_s_array, alpha=0.05, method='bonferroni')
```

```python
bonf_ps = pd.DataFrame(bonferroni_ps[1])
```

```python
##mann_whitney_u test
```

```python
u_s = []
p_s = []
for region in regions:
    case_array = np.array(case_vols[region])
    control_array = np.array(control_vols[region])
    u = [stats.mannwhitneyu(case_array, control_array)[0]]
    u = pd.DataFrame(u)
    u['region']=region
    p = [stats.mannwhitneyu(case_array, control_array)[1]]
    p = pd.DataFrame(p)
    p['region']=region
    u_s.append(u)
    p_s.append(p)
```

```python
u_s = pd.concat(u_s)
p_s = pd.concat(p_s)
```
